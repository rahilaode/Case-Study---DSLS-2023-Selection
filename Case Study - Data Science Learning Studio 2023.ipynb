```python
df2 = pd.read_csv('taxi_zones.csv')
df2.head()
```

```python
df = df.merge(df2[['LocationID', 'borough']], 
         how='inner', left_on='PULocationID', 
         right_on='LocationID').drop(['LocationID'], axis=1).rename(columns={"borough": "pickup"})
df.head()
```

```python
df = df.merge(df2[['LocationID', 'borough']], 
         how='inner', left_on='DOLocationID', 
         right_on='LocationID').drop(['LocationID'], axis=1).rename(columns={"borough": "drop-off"})
df.head()
```

```python
df['route'] = df['pickup'] +' - '+df['drop-off']
```

```python
df.head()
```

```python
df = df[df['pickup'] != 'EWR']
df = df[df['drop-off'] != 'EWR']
df.head()
```

```python
from_ = pd.DataFrame(df['pickup'].value_counts()).reset_index().rename(columns={'index':'borough', 'pickup':'count'})
from_['Pickup/Drop-off'] = 'Pickup'
from_
```

```python
to = pd.DataFrame(df['drop-off'].value_counts()).reset_index().rename(columns={'index':'borough', 'drop-off':'count'})
```

```python
to['Pickup/Drop-off'] = 'Drop-off'
to
```

```python
count = pd.concat([from_, to], ignore_index=True)
count
```

Kemudian kita coba analisis lagi total waktu perjalanan pada saat pickup penumpang pada tiap-tiap borough

```python
pickup_to_dropoff_days = []
pickup_to_dropoff_time = []
for i in df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']:
    pickup_to_dropoff_days.append(int(str(i).split(" ")[0]))
    pickup_to_dropoff_time.append(str(i)[-8:])
```

```python
hours = []
minutes = []
seconds = []
total_time_in_minutes = []
for i in pickup_to_dropoff_time:
    hours.append(int(i.split(':')[0]))
    minutes.append(int(i.split(':')[1]))
    seconds.append(int(i.split(':')[2]))
    #total waktu dalam menit = (jam*60) + menit + (detik/60) 
    total_time_in_minutes.append(round(float((int(i.split(':')[0])*60) 
                                             + int(i.split(':')[1]) 
                                             + (int(i.split(':')[2])/60)), 2))
```

```python
#waktu total perjalanan dalam menit
df['total_time(minute)'] = total_time_in_minutes
df.head()
```

```python
x = 0
y = []
for i in pickup_to_dropoff_days:
    if i>0:
        print("terdapat {} hari perjalanan pada index ke-{}".format(i, x))
        y.append(x)
    x+=1
```

Berdasarkan waktu perjalanan (hari), dapat kita simpulkan bahwa pada 8 baris data tersebut merupakan outlier. Jadi, untuk baris-baris tersebut kita akan hapus

```python
df = df.drop(index=y)
```

```python
plt.figure(figsize=(18,6))
plt.title("Total time from pickupt to drop-off (minutes)", size = 15)
sns.boxplot(data=df, x='pickup', y='total_time(minute)')
plt.xlabel('Borough', size=13)
plt.ylabel('Minutes', size=13)
plt.grid()
plt.show()
```

```python
import numpy as np

print(f'Jumlah baris sebelum memfilter outlier: {len(df)}')

filtered_entries = np.array([True] * len(df))
Q1 = df['total_time(minute)'].quantile(0.25)
Q3 = df['total_time(minute)'].quantile(0.75)
IQR = Q3 - Q1
low_limit = Q1 - (IQR * 1.5)
high_limit = Q3 + (IQR * 1.5)

filtered_entries = ((df['total_time(minute)'] >= low_limit) & (df['total_time(minute)'] <= high_limit)) & filtered_entries

df_cleaned = df[filtered_entries]

print(f'Jumlah baris setelah memfilter outlier: {len(df_cleaned)}')
```

```python
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_format

total_time_value_mean = [df_cleaned['total_time(minute)'][df_cleaned['pickup'] == 'Manhattan'].mean(), 
                         df_cleaned['total_time(minute)'][df_cleaned['pickup'] == 'Queens'].mean(), 
                         df_cleaned['total_time(minute)'][df_cleaned['pickup'] == 'Brooklyn'].mean(), 
                         df_cleaned['total_time(minute)'][df_cleaned['pickup'] == 'Bronx'].mean(), 
                         df_cleaned['total_time(minute)'][df_cleaned['pickup'] == 'Staten Island'].mean()]
plt.figure(figsize=(6,6))
plt.title("")
plt.pie(total_time_value_mean, 
        labels = ['Manhattan', 'Queens', 'Brooklyn', 'Bronx', 'Staten Island'], 
        colors= colors, explode=[0.2, 0, 0, 0, 0], autopct=autopct_format(total_time_value_mean))
plt.show()
```
